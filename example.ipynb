# Canny edge detection implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import scipy as sci
from edge import edge
```

```python
for img_id in [1,2,3]:
    img = cv2.imread(f"demo_images/demo{img_id}.png",cv2.IMREAD_GRAYSCALE)
    img = 255*(img/np.max(img))
    img = img.astype(np.uint8)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 4))
    
    plt.subplot(131)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 255])
    plt.xlabel("x [px]")
    plt.ylabel("y [px]")
    plt.title('image')
    
    plt.subplot(132)
    F = cv2.Canny(img, threshold1 = .2*255, threshold2 = .5*255, apertureSize = 3, L2gradient = False)
    F = F/np.max(F)
    plt.imshow(F, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
    plt.xlabel("x [px]")
    plt.ylabel("y [px]")
    plt.title('cv2 implementation')

    plt.subplot(133)
    F = edge(img, t_low=.2, t_high=.2, sigma = 3)
    plt.imshow(F, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(fraction=0.035, pad=0.04,ticks=[0, 1])
    plt.xlabel("x [px]")
    plt.ylabel("y [px]")
    plt.title('personal implementation')
    plt.rcParams.update({'font.size': 20})
    plt.show()
```
